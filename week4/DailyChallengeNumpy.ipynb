```python
#Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities.
#The temperatures should range from -5 to 35 degrees.
np.random.seed(0)
temperatures = np.random.randint(-5, 36, size=(10, 12))
# print(temperatures)

#Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.

df = pd.DataFrame(temperatures, index=[fake.city() for _ in range(10)])

df.columns = [fake.month_name() for _ in range(12)]
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

df.columns=months

df.head(20)
```

```python
#calculate annual temperature average
df.mean(axis=1)
```

```python
df.mean(axis=1).idxmax()
```

```python
df.mean(axis=1).idxmin()
```

```python
import matplotlib.pyplot as plt

# Transpose the DataFrame to have months as rows and cities as columns
df_transposed = df.T

# Plot the data
df_transposed.plot(kind='line', figsize=(10, 5), title='Temperature in cities')

# Customize the plot
plt.xlabel('Months')
plt.ylabel('Temperature')

# Set x-axis ticks to all month names
plt.xticks(range(len(df.columns)), df.columns, rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
