```python
import pandas as pd
data = pd.read_csv('global_power_plant_database.csv')
data.head()
```

```python
data.info()
```

```python
missing_data=data.isnull().sum()
missing_data.head(50)
```

```python
#impute with median to missing values of generation_gwh_2013 to generation_gwh_2019
data['generation_gwh_2013'].fillna(data['generation_gwh_2013'].median(), inplace=True)
data['generation_gwh_2014'].fillna(data['generation_gwh_2014'].median(), inplace=True)
data['generation_gwh_2015'].fillna(data['generation_gwh_2015'].median(), inplace=True)
data['generation_gwh_2016'].fillna(data['generation_gwh_2016'].median(), inplace=True)
data['generation_gwh_2017'].fillna(data['generation_gwh_2017'].median(), inplace=True)
data['generation_gwh_2018'].fillna(data['generation_gwh_2018'].median(), inplace=True)
data['generation_gwh_2019'].fillna(data['generation_gwh_2019'].median(), inplace=True)
data['estimated_generation_gwh_2013'].fillna(data['estimated_generation_gwh_2013'].median(), inplace=True)
data['estimated_generation_gwh_2014'].fillna(data['estimated_generation_gwh_2014'].median(), inplace=True)
data['estimated_generation_gwh_2015'].fillna(data['estimated_generation_gwh_2015'].median(), inplace=True)
data['estimated_generation_gwh_2016'].fillna(data['estimated_generation_gwh_2016'].median(), inplace=True)
data['estimated_generation_gwh_2017'].fillna(data['estimated_generation_gwh_2017'].median(), inplace=True)



data.drop(['country'], axis=1, inplace=True)
data.drop(['gppd_idnr'], axis=1, inplace=True)
data.drop(['commissioning_year'], axis=1, inplace=True)



data['other_fuel3'].fillna('NA', inplace=True)
data['other_fuel2'].fillna('NA', inplace=True)
data['other_fuel1'].fillna('NA', inplace=True)
data['owner'].fillna('unknown', inplace=True)
data['source'].fillna('unknown', inplace=True)
data['url'].fillna('unknown', inplace=True)
data['geolocation_source'].fillna('unknown', inplace=True)
data['wepp_id'].fillna('unknown', inplace=True)
data['year_of_capacity_data'].fillna('unknown', inplace=True)
data['generation_data_source'].fillna('unknown', inplace=True)


data.info()
```

```python
#Utilize Pandas to summarize key statistics (mean, median, standard deviation) for numerical columns.

#for every column write name of column and type of data
for col in data.columns:
    print(col, data[col].dtype)
```

```python
#Utilize Pandas to summarize key statistics (mean, median, standard deviation) for numerical columns.
#median of the columns with floats or int
median_value = data['estimated_generation_gwh_2013'].median()
print(median_value)
mean_value = data['estimated_generation_gwh_2013'].mean()
print(mean_value)
std_value = data['estimated_generation_gwh_2013'].std()
print(std_value)

#Explore the distribution of power plants by country and fuel type.
d1 = data.groupby(['country_long', 'primary_fuel']).size().reset_index(name='count')
#sort d1 by count in descending order
d1 = d1.sort_values(by='count', ascending=False)
#only take 10 values of most used fuel
d1 = d1.head(10)


import matplotlib.pyplot as plt
import seaborn as sns
#Explore the distribution of power plants by country and fuel type.


sns.barplot(x='country_long', y='count', hue='primary_fuel', data=d1)
plt.xticks(rotation=90)
plt.title('Most Used Fuel by Country')
plt.xlabel('country', fontsize=12)
plt.ylabel('plants count', fontsize=12)
plt.show()

```

```python
#Perform a statistical analysis of power output by fuel type using NumPyâ€™s statistical functions.

fuel_types = data['primary_fuel'].unique()


fuel_statistics = {}

for fuel in fuel_types:
    fuel_data = data[data['primary_fuel'] == fuel]

    generation_columns = [
        'generation_gwh_2013', 'generation_gwh_2014', 'generation_gwh_2015',
        'generation_gwh_2016', 'generation_gwh_2017', 'generation_gwh_2018',
        'generation_gwh_2019'
    ]


    fuel_generation = fuel_data[generation_columns].values

    import numpy as np
    mean_values = np.mean(fuel_generation, axis=0)
    median_values = np.median(fuel_generation, axis=0)
    std_values = np.std(fuel_generation, axis=0)
    min_values = np.min(fuel_generation, axis=0)
    max_values = np.max(fuel_generation, axis=0)


    fuel_statistics[fuel] = {
        'mean': mean_values,
        'median': median_values,
        'std': std_values,
        'min': min_values,
        'max': max_values
    }

for fuel, stats in fuel_statistics.items():
    print(f"Statistics for {fuel}:")
    print(f"Mean Generation: {stats['mean']}")
    print(f"Median Generation: {stats['median']}")
    print(f"Standard Deviation: {stats['std']}")
    print(f"Minimum Generation: {stats['min']}")
    print(f"Maximum Generation: {stats['max']}")
    print("\n")
```
