# **Exercise 1 matrix operations**

```python
import numpy as np
matrix = np.array([[8,2,3],[4,52,6],[7,85,9]])
print(matrix)
```

```python
determinant = (matrix[0,0]) * ((matrix[1,1]) * (matrix[2,2]) - (matrix[1,2]) * (matrix[2,1])) \
            - (matrix[0,1]) * ((matrix[1,0]) * (matrix[2,2]) - (matrix[1,2]) * (matrix[2,0])) \
            + (matrix[0,2]) * ((matrix[1,0]) * (matrix[2,1]) - (matrix[1,1]) * (matrix[2,0]))

print(determinant)
```

```python
inverse_matrix = np.linalg.inv(matrix)
print(inverse_matrix)
```

# **Exercise 6 : Basic Hypothesis Testing**

```python
import numpy as np
from scipy import stats
# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

t_stat, p_value = stats.ttest_rel(productivity_before, productivity_after)

alpha = 0.05
if p_value < alpha:
    print("The training program has a significant effect on employee productivity.")
else:
    print("The training program has no significant effect on employee productivity.")

# if np.median(productivity_after)>np.median(productivity_before):
#     print("The training program has a significant effect on employee productivity.")
# else:
#     print("The training program has no significant effect on employee productivity.")
```

# **Exercise 8 : Time Series Data Manipulation**
