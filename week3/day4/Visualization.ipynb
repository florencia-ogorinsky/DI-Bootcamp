```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Generar datos (edades de 100 estudiantes, entre 18 y 30 años)
np.random.seed(42)  # Para tener resultados reproducibles
edades = np.random.randint(18, 31, size=100)  # 100 estudiantes con edades entre 18 y 30

# Paso 2: Crear el histograma
plt.hist(edades, bins=12, edgecolor='black')  # bins=12 crea 12 grupos de edades
plt.title('Distribución de edades de los estudiantes')
plt.xlabel('Edad')
plt.ylabel('Número de estudiantes')
plt.grid(True)
plt.show()
```

```python
brands = ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Others']
market_share = [40, 30, 15, 10, 5]

plt.pie(market_share, labels=brands, autopct='%1.1f%%')
plt.title('Smartphone Market Share 2023')
plt.show()
```

```python
# 1. Line Plot Exercise:

# Create a line plot for the function y = x^2 for x ranging from 0 to 10. Add appropriate labels and a title.

# np.linspace(start, stop, num) es una función de NumPy que genera un array de números espaciados uniformemente entre el valor start (inicio) y stop (final), con un total de num elementos.
# En este caso:
# start = 0: el primer número en el array será 0.
# stop = 10: el último número en el array será 10.
# num = 100: habrá 100 valores en el array entre 0 y 10 (inclusive).

x= np.linspace(0,10,100)
y= x**2
#y is also array
plt.plot(x, y, color='green',  linewidth=2)
plt.show()
```

# **Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample data
data = sns.load_dataset('tips')
sns.histplot(data['total_bill'], bins=20, color='skyblue')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
data.head()
```

```python
sns.kdeplot(data['total_bill'], shade=True, color='purple')
plt.title('KDE of Total Bill')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='scatter', color='green')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
sns.barplot(x='class', y='survived', data=titanic, palette='Blues')
plt.title('Survival Rate by Class')
plt.show()
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='Set3')
plt.title('Age Distribution by Class')
plt.show()
```

```python
sns.violinplot(x='class', y='age', data=titanic, palette='muted')
plt.title('Age Distribution by Class')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=data, aspect=1.5, height=5)
plt.title('Scatter Plot with Regression Line')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='reg', color='orange')
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species', palette='Dark2')
plt.show()
```

```python
correlation_matrix = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr() # Select only numerical columns for correlation

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
data = sns.load_dataset('tips')
g = sns.FacetGrid(data, col='time', row='smoker', margin_titles=True)
g.map(sns.scatterplot, 'total_bill', 'tip', color='red')
plt.show()
```

```python
iris = sns.load_dataset('iris')
g = sns.PairGrid(iris, hue='species')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap='Blues')
g.map_diag(sns.histplot, kde=True, color='teal')
g.add_legend()
plt.show()
```
