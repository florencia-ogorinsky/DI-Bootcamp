```python
import pandas as pd
data = pd.read_csv('2019.csv')
data.head()
data.info()
data.describe(include='all')
#there are 156 values in each. Dont need to adjust missing values.
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['Social support'], data['Score'])
plt.xlabel('Social Support')
plt.ylabel('Score')
plt.title('Social Support vs Score')
plt.show()
```

```python
import matplotlib.pyplot as plt
#compare GDP per Capita and Healthy Life Expectancy across different country.


import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(50, 10))


ax1.bar(data['Country or region'], data['Healthy life expectancy'], color='green')
ax1.set_title('Life Expectancy by Country')
ax1.set_xlabel('Country or Region')
ax1.set_ylabel('Life Expectancy')
ax1.tick_params(axis='x', rotation=90)  # Rotar las etiquetas del eje x

# Segundo subgrfico (ax2) para GDP per Capita
ax2.bar(data['Country or region'], data['GDP per capita'], color='blue')
ax2.set_title('GDP per Capita by Country')
ax2.set_xlabel('Country or Region')
ax2.set_ylabel('GDP per Capita')
ax2.tick_params(axis='x', rotation=90)  # Rotar las etiquetas del eje x

# Mostrar los subgrficos
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(50, 10))

ax1.bar(data['Country or region'], data['GDP per capita'], color='blue')
ax1.set_xlabel('Country or Region')
ax1.set_ylabel('GDP per Capita', color='blue')
ax1.tick_params(axis='x', rotation=90)


ax2 = ax1.twinx()
ax2.plot(data['Country or region'], data['Healthy life expectancy'])
ax2.set_ylabel('Healthy Life Expectancy', color='green')


ax1.set_title('GDP per Capita and Healthy Life Expectancy by Country')

plt.tight_layout()
plt.show()
```
