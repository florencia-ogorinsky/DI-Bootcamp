# **The Pareto Principle**

# Pareto for products and sales

```python
data=pd.read_csv('US Superstore data.csv')
#group data by product id, reset index
data=data.groupby('Product ID').sum().reset_index()
data=data.sort_values(by='Sales',ascending=False)


#sum amounts from 'Sales'
total_sales=data['Sales'].sum()

#calculate 80% of total_sales
target_sales=total_sales*0.8


def get_product():
    sales_sum = 0
    product_list = []
    for index, row in data.iterrows():
        sales_sum += row['Sales']
        product_list.append(row['Product ID'])
        if sales_sum >= target_sales:
            break
    return product_list


products=pd.Series(get_product()).unique()
print(products)


data_modified=data[data['Product ID'].isin(products)].head(10)

import matplotlib.pyplot as plt
plt.pie(data_modified['Sales'], labels=data_modified['Product ID'], autopct='%1.1f%%')
plt.title('Sales by Product')
plt.show()


#count how many rows the data has
#The percentage of prod which contributes to 80% of sales is:
print((len(products)/len(data))*100,'%')
```

# Pareto for customers and profit

```python
data=pd.read_csv('US Superstore data.csv')
#group data by Customer ID, reset index
data=data.groupby('Customer ID').sum().reset_index()
data=data.sort_values(by='Profit',ascending=False)


#sum amounts from 'profit'
total_Profit=data['Profit'].sum()

#calculate 80% of total_Profit
target_Profit=total_Profit*0.8


def get_customer():
    Profit_sum = 0
    customer_list = []
    for index, row in data.iterrows():
        Profit_sum += row['Profit']
        customer_list.append(row['Customer ID'])
        if Profit_sum >= target_Profit:
            break
    return customer_list


customers=pd.Series(get_customer()).unique()
print(customers)

data_modified2=data[data['Customer ID'].isin(customers)].head(10)

import matplotlib.pyplot as plt
plt.pie(data_modified2['Profit'], labels=data_modified2['Customer ID'], autopct='%1.1f%%')
plt.title('Profit by Customer')
plt.show()

#The percentage of customers which contributes to 80% of profit is:
print((len(customers)/len(data))*100,'%')
```

# **Can we apply Pareto principle to customers and Sales ?**

```python
data=pd.read_csv('US Superstore data.csv')
#group data by Customer ID, reset index
data=data.groupby('Customer ID').sum().reset_index()
data=data.sort_values(by='Sales',ascending=False)


#sum amounts from 'sales'
total_Sales=data['Sales'].sum()

#calculate 80% of total_Sales
target_Sales=total_Sales*0.8


def get_customer():
    Sales_sum = 0
    customer_list = []
    for index, row in data.iterrows():
        Sales_sum += row['Sales']
        customer_list.append(row['Customer ID'])
        if Sales_sum >= target_Sales:
            break
    return customer_list


customers=pd.Series(get_customer()).unique()
print(customers)

data_modified2=data[data['Customer ID'].isin(customers)].head(10)

import matplotlib.pyplot as plt
plt.pie(data_modified2['Sales'], labels=data_modified2['Customer ID'], autopct='%1.1f%%')
plt.title('Sales by Customer')
plt.show()

#The percentage of customers which contributes to 80% of Sales is:
print((len(customers)/len(data))*100,'%')
```

# **Based on the analysis, make decisions on which states and cities to prioritize for marketing strategies**

```python
import pandas as pd
data = pd.read_csv('US Superstore data.csv')
data = data.groupby('State').sum().reset_index()
data = data.sort_values(by='Profit', ascending=False)
total_Profit = data['Profit'].sum()

target_Profit = total_Profit * 0.8

def get_state():
    Profit_sum = 0
    state_list = []
    for index, row in data.iterrows():
        Profit_sum += row['Profit']
        state_list.append(row['State'])
        if Profit_sum >= target_Profit:
            break
    return state_list


states = pd.Series(get_state()).unique()
# print(states)

#states not in pareto:
data_modified3 = data[~data['State'].isin(states)].head(10)
data_modified3 = data_modified3.sort_values(by='Profit', ascending=True)

print(data_modified3['State'], "This are states with lower profit and outside of the Pareto, we need to reforce marketing")
```
