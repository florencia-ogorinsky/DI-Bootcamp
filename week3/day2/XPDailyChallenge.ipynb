```python
from scipy.ndimage import rotate
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt


def rotate_image_90_degrees(image):
    return rotate(image, 90, reshape=True, order=1)

def flip_image(image):
    flipped_horizontally = ImageOps.mirror(image)  # Flip horizontally
    flipped_vertically = ImageOps.flip(flipped_horizontally)  # Flip vertically
    return flipped_vertically


def zoom_in(image):
    width, height = image.size
    new_size = (int(width * 1.2), int(height * 1.2))
    zoomed_image = image.resize(new_size)
    return zoomed_image

image_path = 'flower_images/flower_images/0001.png'
original_image = Image.open(image_path)

image_array = np.array(original_image)

rotated_image = rotate_image_90_degrees(image_array)
flipped_image = flip_image(original_image)
zoomed_image = zoom_in(original_image)

plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.imshow(rotated_image)
plt.title("Rotated Image")

plt.subplot(2, 2, 2)
plt.imshow(flipped_image)
plt.title("Flipped Image")

plt.subplot(2, 2, 3)
plt.imshow(zoomed_image)
plt.title("Zoomed Image")

plt.show()
```
