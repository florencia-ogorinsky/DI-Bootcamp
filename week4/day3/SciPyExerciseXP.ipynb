```python
import numpy as np
from scipy import stats
data = [12, 15, 13, 12, 18, 20, 22, 21]
mean_scipy=stats.tmean(data)
print(mean_scipy)
median_scipy=np.median(data)
print(median_scipy)
variance_scipy=stats.tvar(data)
print(variance_scipy)
std_scipy=stats.tstd(data)
print(std_scipy)
```

```python
house_sizes= [50, 70, 80, 100, 120]
house_prices= [150000, 200000, 210000, 250000, 280000]

slope,intercept,r_value,p_value,std_err=stats.linregress(house_sizes,house_prices)

print("Slope:", slope)
print("Intercept:", intercept)

#predicted price for house of 90 meters:
#y = mx + b
predicted_price=slope*90+intercept
print("Predicted price:", predicted_price)

import matplotlib.pyplot as plt
plt.scatter(house_sizes, house_prices)
plt.plot(house_sizes, slope*np.array(house_sizes)+intercept, color='red')



# slope indicates how much increase the price (1828$) for each more meter
# intercept is if the house would have 0 mtrs would cost 64383$
```

# **Exercise 6**

```python
fertilizer_1= [5, 6, 7, 6, 5]
fertilizer_2= [7, 8, 7, 9, 8]
fertilizer_3= [4, 5, 4, 3, 4]

f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("F-value:", f_value)
print("P-value:", p_value)

#F is the variance between the groups is more than variance inside each of them
#P more than 0.05 means there are no significant variance between them
```

# **Exercise 7**

```python
# Example: Binomial Distribution
# getting exactly 5 heads in 10 coin flips.


n, p = 10, 0.5  # number of trials, probability of each trial (0.5 is half of possibilitird, cause 5 is half of 10)
binom_dist = stats.binom(n, p)

number_successes = np.arange(0, n+1)
pmf_values = binom_dist.pmf(number_successes)
plt.figure(figsize=(6, 4))
plt.bar(number_successes, pmf_values)
plt.title('Binomial Distribution\n$n=10$, $p=0.5$')
plt.xlabel('Number of Successes')
plt.ylabel('PMF')
plt.show()


```

# **Exercise 8**

```python
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

from scipy.stats import pearsonr

# Calculate Pearson Correlation between Monthly Income and age
corr, _ = pearsonr(data['income'], data['age'])
print(f'Pearson Correlation Coefficient: {corr:.2f}')

#1 indicates a perfect positive linear relationship. direct relation between variables

#Spearman
from scipy.stats import spearmanr
corr, _ = spearmanr(data['income'], data['age'])
print(f'Spearman Correlation Coefficient: {corr:.2f}')

# 1 indicates they both move at a constant rate
```
