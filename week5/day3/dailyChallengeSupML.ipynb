```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
                df.drop('diagnosis', axis=1),
                df['diagnosis'],
                test_size=0.2,
                random_state=42)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

# **We used standard scaler because we want to use K Nearest Neighboars and we dont want the features to have very different scales**

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
predictions1 = logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print("Confusion Matrix: \n", confusion_matrix(y_test, predictions1))
print('\n')
print(classification_report(y_test, predictions1))
```

#**The model has high precission to determinate both M and B. For class M precission is 1 which means is always predicting good when is malign which would be the most important. In recall we see that there is a small percentage of Malign that the model is not predicting good (0.95). f1 is like an average between the other 2. In support we can see what we see also in the confussion matrix, from 114 patients, 71 were B and predicted as B, 2 were M and predicted as B (here the model failed), no people with B were predicted as M, and 41 were M and also predicted as M**

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=100000)

model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

class_names = ['B', 'M']

matrix = confusion_matrix(y_test, pred)

dataframe = pd.DataFrame(matrix, index=class_names, columns=class_names)

sns.heatmap(dataframe, annot=True, cbar=None, cmap="Blues", fmt='g')

plt.title("Confussion matrix")
plt.tight_layout()
plt.ylabel("True")
plt.xlabel("Predicted")

plt.show()

print(classification_report(y_test, pred))
```

```python
from sklearn.metrics import accuracy_score

logreg_acc = accuracy_score(y_test, predictions1)
print("accuracy LR: ", logreg_acc)
```

# **The percentage of predicted correctly divided total predicted is very high which is good, but is not the best metric in a case where there might be more with B than M, and the target variable is binary**

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

error_rate = [np.mean(KNeighborsClassifier(n_neighbors=i).fit(X_train, y_train).predict(X_test) != y_test) for i in range(1, 42)]

plt.figure(figsize=(12,6))
plt.plot(range(1, 42), error_rate, color='red',marker='o', markerfacecolor='b')
plt.xticks(np.arange(1, 42))
plt.title('Error Rate vs K-value')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Error Rate')
plt.show()
```

# **When number of neighbors is 9,34,35,36,40 and 41 there is less error rate.**

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train, y_train)
predictions2 = knn.predict(X_test)

print(confusion_matrix(y_test, predictions2))
print("\n")
print(classification_report(y_test, predictions2))

knn_model_acc = accuracy_score(y_test, predictions2)
print("Accuracy of K Neighbors Classifier Model is: ", knn_model_acc)
```

# **Using number of neighbors=9 because there is a good balance between precission and error rate. We can see that the correct number of predictions is now less (now is 69 when before it was 71, now the number of people with B that was predicted as M is 2 when before was 0), and the rest is the same. Because of these reasons we can say KNN is less efficient. We will see the same results in the graph below**

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train, y_train)
predictions2 = knn.predict(X_test)

class_names = ['B', 'M']

matrix = confusion_matrix(y_test, predictions2)

dataframe = pd.DataFrame(matrix, index=class_names, columns=class_names)

sns.heatmap(dataframe, annot=True, cbar=None, cmap="Blues", fmt='g')

plt.title("Confusion Matrix for KNN (n_neighbors=9)")
plt.tight_layout()
plt.ylabel("True Class")
plt.xlabel("Predicted Class")

plt.show()

print(classification_report(y_test, predictions2))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
predictions4 = rfc.predict(X_test)

print("Confusion Matrix: \n", confusion_matrix(y_test, predictions4))
print("\n")
print(classification_report(y_test, predictions4))

rfc_acc = accuracy_score(y_test, predictions4)
print("Accuracy : ", rfc_acc)
```

```python
from sklearn.svm import SVC

svc_model = SVC(kernel="rbf")
svc_model.fit(X_train, y_train)
predictions5 = svc_model.predict(X_test)

print("Confusion Matrix: \n", confusion_matrix(y_test, predictions5))
print("\n")
print(classification_report(y_test, predictions5))

svm_acc = accuracy_score(y_test, predictions5)
print("accuracy SVM: ", svm_acc)
```

# **The accuracy in SMV (0.98) and Linear Regression (also 0.98) is the greatest, also the people correctly predicted are the same 71 in both, so we can say these 2 are the best models to predict our target variable**
